#### Interactive Checks

```python
p.returns[p.assets].head()
```

```python
er = p.returns[p.assets].mean() * 252
er
```

```python
np.dot(p.weights, er)
```

```python
cov = p.returns[p.assets].cov() * 252  # covariance matrix
cov
```

```python
np.dot(p.weights, np.dot(cov, p.weights))  # portfolio variance
```

```python
np.dot(p.weights, np.dot(cov, p.weights)) ** 0.5  # portfolio volatility
```

```python
p.weights @ cov @ p.weights
```
