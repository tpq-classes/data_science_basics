## Socket Connection

```python
import zmq
```

```python
context = zmq.Context()
```

```python
socket = context.socket(zmq.SUB)
```

```python
socket.connect('tcp://127.0.0.1:5555')
```

```python
socket.setsockopt_string(zmq.SUBSCRIBE, '')
```

## Retrieving & Visualizing Tick Data

```python
import plotly.graph_objects as go
```

```python
fig = go.FigureWidget()
fig.add_scatter(name='AAPL')
fig.add_scatter(name='MSFT')
fig
```

```python
times_0 = list()
prices_0 = list()
times_1= list()
prices_1 = list()
```

```python
i = 0
while i < 25:
    i += 1
    msg = socket.recv_pyobj()
    if msg['SYMBOL'] == 'AAPL':
        times_0.append(msg['TIME'])
        prices_0.append(msg['PRICE'])
        fig.data[0].x = times_0
        fig.data[0].y = prices_0
    elif msg['SYMBOL'] == 'MSFT':
        times_1.append(msg['TIME'])
        prices_1.append(msg['PRICE'])
        fig.data[1].x = times_1
        fig.data[1].y = prices_1
```

```python
fig = go.FigureWidget()
fig.add_scatter(name='AAPL')
fig.add_scatter(name='SMA1')
fig.add_scatter(name='SMA2')
fig
```

```python
ticks = pd.DataFrame()
```

```python
i = 0
while i < 100:
    i += 1
    msg = socket.recv_pyobj()
    df = pd.DataFrame(msg, index=[0])
    ticks = pd.concat((ticks, df), ignore_index=True)
    aapl = ticks[ticks['SYMBOL'] == 'AAPL']
    aapl = aapl.set_index('TIME')
    aapl['SMA1'] = aapl['PRICE'].rolling(3).mean()
    aapl['SMA2'] = aapl['PRICE'].rolling(7).mean()
    fig.data[0].x = aapl.index
    fig.data[0].y = aapl.PRICE
    fig.data[1].x = aapl.index
    fig.data[1].y = aapl.SMA1
    fig.data[2].x = aapl.index
    fig.data[2].y = aapl.SMA2
```
