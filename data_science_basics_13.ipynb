## Financial Data

```python
url = 'https://certificate.tpq.io/rl4finance.csv'
```

```python
raw = pd.read_csv(url, index_col=0, parse_dates=True).dropna()
```

```python
raw.info()
```

## Time Series Visualization

```python
raw.plot();
```

```python
(raw / raw.iloc[0]).plot(lw=1);
```

```python
symbols = ['.SPX', 'SPY']
```

```python
data = raw[symbols].copy()
```

```python
data.plot();
```

```python
data.plot(subplots=True);
```

```python
data.plot(secondary_y='SPY');
```

```python
(data / data.iloc[0]).plot();
```

## Return Visualizations

```python
rets = raw.pct_change()  # daily simple returns
```

```python
rets.head().round(4)
```

```python
print(rets.head().round(4))
```

```python
rets['.SPX'].plot();
```

```python
rets['.SPX'].hist(bins=50);
```

```python
# rets.plot?
```

```python
rets.hist(bins=50, figsize=(10, 10));
```

## Two Return Series Scatter Plot

```python
symbols = ['.SPX', '.VIX']
```

```python
raw[symbols].plot(subplots=True);
```

```python
raw[symbols].plot(secondary_y='.VIX', lw=1);
```

```python
rets[symbols].plot(alpha=0.7);
```

```python
rets[symbols].plot(kind='scatter', x='.SPX', y='.VIX', alpha=0.7);
```

```python
rets[symbols].plot(kind='scatter', x='.SPX', y='.VIX',
                   c=rets.index, alpha=0.7);
```

```python
rets[symbols].plot(kind='scatter', x='.SPX', y='.VIX',
                   c=rets.index, alpha=0.7, cmap='coolwarm');
```

```python
c = (rets.index - rets.index[0]).days
```

```python
c
```

```python
rets[symbols].plot(kind='scatter', x='.SPX', y='.VIX',
                   c=c, alpha=0.7, cmap='Reds',
                  colorbar=True);
```
