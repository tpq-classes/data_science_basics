## Financial Data

```python
url = 'https://certificate.tpq.io/rl4finance.csv'
```

```python
raw = pd.read_csv(url, index_col=0, parse_dates=True).dropna()
```

```python
raw.info()
```

## Standard Plots

```python
(raw / raw.iloc[0]).plot(lw=1);
```

```python
raw.plot(lw=1, subplots=True, figsize=(10, 12));
```

```python
rets = raw.pct_change()
```

```python
rets.hist(figsize=(10, 10), bins=50);
```

```python
rets['AAPL.O'].hist(bins=50);
```

## Seaborn Plots

```python
import seaborn as sns
```

```python
# sns.histplot?
```

```python
sns.histplot(rets['AAPL.O'], bins=50, stat='probability');
```

```python
rets['NORMAL'].describe()
```

```python
sns.histplot(rets['NORMAL'], bins=50, stat='probability', kde=True);
```

```python
sns.histplot(rets[['AAPL.O', 'NORMAL']], bins=50, stat='probability', kde=True);
```

```python
sns.kdeplot(rets[['AAPL.O', 'NORMAL']]);
```

```python
sns.ecdfplot(rets[['AAPL.O', 'NORMAL']]);
```

```python
sns.pairplot(rets[['AAPL.O', '.SPX']]);
```

```python
sns.pairplot(rets[['AAPL.O', '.SPX']], diag_kind='kde');
```

```python
sns.jointplot(rets[['AAPL.O', '.SPX']], x='.SPX', y='AAPL.O');
```

## Interactive Plotting with `cufflinks`

```python
symbols = ['AAPL.O', '.SPX']
```

```python
data = pd.DataFrame(raw[symbols])
data.info()
```

```python
data.plot(subplots=True);
```

```python
import cufflinks as cf
cf.go_offline()
import plotly.io as pio
pio.renderers.default = 'colab'
import warnings as w
w.simplefilter('ignore')
```

```python
data.iplot(subplots=True)
```

```python
data.normalize().iplot()
```

```python
rets[symbols].iplot();
```

```python
rets[symbols].iplot(kind='hist');
```

```python
rets[symbols].iplot(kind='hist', subplots=True);
```

```python
data.normalize().iplot(theme='solar')
```

```python
rets.iplot(kind='hist', subplots=True, dimensions=(1028, 768));
```

```python
rets.corr().round(3)
```

```python
rets.corr().iplot(kind='heatmap', colorscale='reds')
```

```python
rets.corr().iplot(kind='heatmap', colorscale='rdbu')
```
