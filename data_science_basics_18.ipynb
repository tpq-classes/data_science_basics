## Reading & Visualizing Tick Data

```python
data = pd.read_hdf('ticks.h5')
```

```python
data.index = pd.DatetimeIndex(data.index)
```

```python
data.info()
```

```python
data[data['SYMBOL'] == 'AAPL'].plot();
```

```python
data[data['SYMBOL'] == 'MSFT'].plot();
```

```python
ax = data[data['SYMBOL'] == 'AAPL'].plot()
data[data['SYMBOL'] == 'MSFT'].plot(ax=ax);
```

```python
data_ = pd.DataFrame(data[data['SYMBOL'] == 'AAPL']['PRICE'])
data_.columns = ['AAPL']
```

```python
data_.info()
```

```python
# data_.join?
```

```python
data_ = data_.join(data[data['SYMBOL'] == 'MSFT']['PRICE'], how='outer')
```

```python
data_.columns = ['AAPL', 'MSFT']
```

```python
data_.head(7)
```

```python
data_.plot();
```

```python
data__ = data_.resample('5s', label='right').last().ffill()
```

```python
data__.info()
```

```python
data__.head()
```

```python
data__.plot();
```
