## First Visualizations

```python
plt.plot(l);
```

```python
x = list(range(100, 126))
```

```python
plt.plot(x, l);
```

```python
def f(x):
    return math.log(x)
```

```python
x = range(1, 26)
```

```python
y = [f(e) for e in x]
```

```python
plt.plot(x, y);
```

## Random Walks

```python
rn = np.random.standard_normal(252)  # legacy way of generating (pseudo-)random number
```

```python
rng = np.random.default_rng(100)
```

```python
rn = rng.standard_normal(252)
```

```python
np.set_printoptions(suppress=True)
```

```python
rn[0] = 0
```

```python
rn[:10]
```

```python
plt.plot(rn);
```

```python
plt.plot(rn.cumsum());
```

```python
plt.plot(rn.cumsum() + 100);
```

```python
rn = rng.standard_normal((252, 5))
```

```python
rn.shape
```

```python
rn
```

```python
rn[0] = 0
```

```python
plt.plot(rn);
```

```python
plt.plot(rn.cumsum());
```

```python
plt.plot(rn.cumsum(axis=0));
```

```python
plt.plot(rn.cumsum(axis=0) + 100);
```

```python
rw = pd.DataFrame(rn.cumsum(axis=0) + 100)
```

```python
rw.head()
```

```python
plt.plot(rw);
```

```python
rw.columns = list('ABCDE')
```

```python
rw.plot();
```

```python
rw.plot(style=['--', 'o', '.', '-.', '--']);
```

```python
rw.plot(style=['g--', 'bo', 'r.', 'r-.', 'r--'], title='Random Walks');
```

```python
index = pd.date_range('2025-7-1', periods=len(rw), freq='B')
```

```python
index
```

```python
rw.index = index
```

```python
rw.plot();
```
